```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "data/mouse_drug_data.csv"
clinical_trial_data_to_load = "data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
mouse_drug_data_df=pd.read_csv(mouse_drug_data_to_load)
clinicaltrial_data_df=pd.read_csv(clinical_trial_data_to_load)

# Combine the data into a single dataset
merged_df=pd.merge(clinicaltrial_data_df,mouse_drug_data_df,how="left",on=["Mouse ID", "Mouse ID"])

# Display the data table for preview
merged_df
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
mean_tumor_vol=merged_df.groupby(["Drug", "Timepoint"]).mean()["Tumor Volume (mm3)"]

# Convert to DataFrame
mean_tumor_df=pd.DataFrame(mean_tumor_vol).reset_index()

# Preview DataFrame
mean_tumor_df
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
se_tumor_vol=merged_df.groupby(["Drug", "Timepoint"]).sem()["Tumor Volume (mm3)"]

# Convert to DataFrame
se_tumor_df=pd.DataFrame(se_tumor_vol).reset_index()

# Preview DataFrame
se_tumor_df
```

```python
# Minor Data Munging to Re-Format the Data Frames
munge_mean_tumor_df=mean_tumor_df.pivot(index="Timepoint",columns="Drug",values="Tumor Volume (mm3)")
munge_se_tumor_df=se_tumor_df.pivot(index="Timepoint",columns="Drug",values="Tumor Volume (mm3)")

# Preview that Reformatting worked
munge_mean_tumor_df
```

```python
# Generate the Plot (with Error Bars)
plt.errorbar(munge_mean_tumor_df.index,y=munge_mean_tumor_df["Capomulin"],yerr=munge_se_tumor_df["Capomulin"], 
             linewidth=1,marker="o",markersize=3,linestyle='--',label="Capomulin")
plt.errorbar(munge_mean_tumor_df.index,y=munge_mean_tumor_df["Ceftamin"],yerr=munge_se_tumor_df["Ceftamin"], 
             linewidth=1,marker="o",markersize=3,linestyle='--',label="Ceftamin")
plt.errorbar(munge_mean_tumor_df.index,y=munge_mean_tumor_df["Infubinol"],yerr=munge_se_tumor_df["Infubinol"], 
             linewidth=1,marker="o",markersize=3,linestyle='--',label="Infubinol")
plt.errorbar(munge_mean_tumor_df.index,y=munge_mean_tumor_df["Ketapril"],yerr=munge_se_tumor_df["Ketapril"], 
             linewidth=1,marker="o",markersize=3,linestyle='--',label="Ketapril")
plt.errorbar(munge_mean_tumor_df.index,y=munge_mean_tumor_df["Naftisol"],yerr=munge_se_tumor_df["Naftisol"], 
             linewidth=1,marker="o",markersize=3,linestyle='--',label="Naftisol")
plt.errorbar(munge_mean_tumor_df.index,y=munge_mean_tumor_df["Placebo"],yerr=munge_se_tumor_df["Placebo"], 
             linewidth=1,marker="o",markersize=3,linestyle='--',label="Placebo")
plt.errorbar(munge_mean_tumor_df.index,y=munge_mean_tumor_df["Propriva"],yerr=munge_se_tumor_df["Propriva"], 
             linewidth=1, marker="o", markersize=3, linestyle='--',label="Propriva")
plt.errorbar(munge_mean_tumor_df.index,y=munge_mean_tumor_df["Ramicane"],yerr=munge_se_tumor_df["Ramicane"], 
             linewidth=1,marker="o",markersize=3,linestyle='--',label="Ramicane")
plt.errorbar(munge_mean_tumor_df.index,y=munge_mean_tumor_df["Stelasyn"],yerr=munge_se_tumor_df["Stelasyn"], 
             linewidth=1,marker="o",markersize=3,linestyle='--',label="Stelasyn")
plt.errorbar(munge_mean_tumor_df.index,y=munge_mean_tumor_df["Zoniferol"],yerr=munge_se_tumor_df["Zoniferol"], 
             linewidth=1,marker="o",markersize=3,linestyle='--',label="Zoniferol")

# Save the Figure
plt.title("Tumor Response to Treatment Overtime")
plt.xlabel("Time (Days)")
plt.ylabel("Tumor Size (mm3)")
plt.xticks(np.arange(0, 50, 5))
plt.grid()
plt.legend(loc="upper right", bbox_to_anchor=(1.3,1))
plt.savefig("Fig1.png")

# Show the Figure
plt.show()
```

## Metastatic Response to Treatment

```python
# Store the Mean Met. Site Data Grouped by Drug and Timepoint 
mean_met_site=merged_df.groupby(["Drug", "Timepoint"]).mean()["Metastatic Sites"]

# Convert to DataFrame
mean_met_site_df=pd.DataFrame(mean_met_site).reset_index()

# Preview DataFrame
mean_met_site_df
```

```python
# Store the Standard Error associated with Met. Sites Grouped by Drug and Timepoint 
se_met_site=merged_df.groupby(["Drug", "Timepoint"]).sem()["Metastatic Sites"]

# Convert to DataFrame
se_met_site_df=pd.DataFrame(se_met_site).reset_index()

# Preview DataFrame
se_met_site_df
```

```python
# Minor Data Munging to Re-Format the Data Frames
munge_mean_met_site_df=mean_met_site_df.pivot(index="Timepoint", columns="Drug", values="Metastatic Sites")
munge_se_met_site_df=se_met_site_df.pivot(index="Timepoint", columns="Drug", values="Metastatic Sites")

# Preview that Reformatting worked
munge_mean_met_site_df
```

```python
# Generate the Plot (with Error Bars)
plt.errorbar(munge_mean_met_site_df.index,y=munge_mean_met_site_df["Capomulin"],yerr=munge_se_met_site_df["Capomulin"], 
             linewidth=1,marker="o",markersize=3,linestyle='--',label="Capomulin")
plt.errorbar(munge_mean_met_site_df.index,y=munge_mean_met_site_df["Ceftamin"],yerr=munge_se_met_site_df["Ceftamin"], 
             linewidth=1,marker="o",markersize=3,linestyle='--',label="Ceftamin")
plt.errorbar(munge_mean_met_site_df.index,y=munge_mean_met_site_df["Infubinol"],yerr=munge_se_met_site_df["Infubinol"], 
             linewidth=1,marker="o",markersize=3,linestyle='--',label="Infubinol")
plt.errorbar(munge_mean_met_site_df.index,y=munge_mean_met_site_df["Ketapril"],yerr=munge_se_met_site_df["Ketapril"], 
             linewidth=1,marker="o",markersize=3,linestyle='--',label="Ketapril")
plt.errorbar(munge_mean_met_site_df.index,y=munge_mean_met_site_df["Naftisol"],yerr=munge_se_met_site_df["Naftisol"], 
             linewidth=1,marker="o",markersize=3,linestyle='--',label="Naftisol")
plt.errorbar(munge_mean_met_site_df.index,y=munge_mean_met_site_df["Placebo"],yerr=munge_se_met_site_df["Placebo"], 
             linewidth=1,marker="o",markersize=3,linestyle='--',label="Placebo")
plt.errorbar(munge_mean_met_site_df.index,y=munge_mean_met_site_df["Propriva"],yerr=munge_se_met_site_df["Propriva"], 
             linewidth=1,marker="o",markersize=3,linestyle='--',label="Propriva")
plt.errorbar(munge_mean_met_site_df.index,y=munge_mean_met_site_df["Ramicane"],yerr=munge_se_met_site_df["Ramicane"], 
             linewidth=1,marker="o",markersize=3,linestyle='--',label="Ramicane")
plt.errorbar(munge_mean_met_site_df.index,y=munge_mean_met_site_df["Stelasyn"],yerr=munge_se_met_site_df["Stelasyn"], 
             linewidth=1,marker="o",markersize=3,linestyle='--',label="Stelasyn")
plt.errorbar(munge_mean_met_site_df.index,y=munge_mean_met_site_df["Zoniferol"],yerr=munge_se_met_site_df["Zoniferol"], 
             linewidth=1,marker="o",markersize=3,linestyle='--',label="Zoniferol")

# Save the Figure
plt.title("Metastic Spread During Treatment Over Time")
plt.xlabel("Time (Days)")
plt.ylabel("Number of Metastic Sites")
plt.xticks(np.arange(0, 50, 5))
plt.grid()
plt.legend(loc="upper right", bbox_to_anchor=(1.3,1))
plt.savefig("Fig2.png")

# Show the Figure
plt.show()
```

## Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)
count_mice=merged_df.groupby(["Drug", "Timepoint"]).count()["Mouse ID"]

# Convert to DataFrame
count_mice_df=pd.DataFrame(count_mice).reset_index()

# Preview DataFrame
count_mice_df
```

```python
# Minor Data Munging to Re-Format the Data Frames
munge_count_mice_df=count_mice_df.pivot(index="Timepoint", columns="Drug", values="Mouse ID")

# Preview the Data Frame
munge_count_mice_df
```

```python
# Generate the Plot (Accounting for percentages)
count_mice_df2= count_mice_df.pivot(index="Timepoint", columns="Drug")["Mouse ID"]
count_mice_df2
percentage_survival_rate=round(count_mice_df2/(count_mice_df2.iloc[0])*100, 2)
percentage_survival_rate

plt.plot(percentage_survival_rate.index,percentage_survival_rate["Capomulin"], 
             linewidth=1,marker="o",markersize=3,linestyle='--',label="Capomulin")
plt.plot(percentage_survival_rate.index,percentage_survival_rate["Ceftamin"], 
             linewidth=1,marker="o",markersize=3,linestyle='--',label="Ceftamin")
plt.plot(percentage_survival_rate.index,percentage_survival_rate["Infubinol"],
             linewidth=1,marker="o",markersize=3,linestyle='--',label="Infubinol")
plt.plot(percentage_survival_rate.index,percentage_survival_rate["Ketapril"],
             linewidth=1,marker="o",markersize=3,linestyle='--',label="Ketapril")
plt.plot(percentage_survival_rate.index, percentage_survival_rate["Naftisol"],
             linewidth=1,marker="o",markersize=3,linestyle='--',label="Naftisol")
plt.plot(percentage_survival_rate.index,percentage_survival_rate["Placebo"],
             linewidth=1,marker="o",markersize=3,linestyle='--',label="Placebo")
plt.plot(percentage_survival_rate.index,percentage_survival_rate["Propriva"],
             linewidth=1,marker="o",markersize=3,linestyle='--',label="Propriva")
plt.plot(percentage_survival_rate.index,percentage_survival_rate["Ramicane"],
             linewidth=1,marker="o",markersize=3,linestyle='--',label="Ramicane")
plt.plot(percentage_survival_rate.index,percentage_survival_rate["Stelasyn"],
             linewidth=1,marker="o",markersize=3,linestyle='--',label="Stelasyn")
plt.plot(percentage_survival_rate.index,percentage_survival_rate["Zoniferol"],
             linewidth=1,marker="o",markersize=3,linestyle='--',label="Zoniferol")

# Save the Figure
plt.title("Mice Survival Rate over Time")
plt.xlabel("Treatment Duration (Days)")
plt.ylabel("Survival Rate (%)")
plt.xticks(np.arange(0, 50, 5))
plt.grid()
plt.legend(loc="upper right", bbox_to_anchor=(1.3,1))
plt.savefig("Fig3.png")

# Show the Figure
plt.show()
```
